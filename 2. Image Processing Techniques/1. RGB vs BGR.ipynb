```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

## Loading an image using OpenCV

```python
img = cv.imread('rgb.jpeg')
plt.imshow(img)
```

## Converting BGR to RGB

```python
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
```
